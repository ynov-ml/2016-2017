Nous constatons que les ages ne sont pas tous remplis. Nous allons donc remplacer les âges où la valeur est non renseignée par la médiane . 

```python
dfTrain.Age.median()
```

Calculons maintenant la médiane par genre et par classe. On commence par créer une colonne Genre que nous allons remplir. 

```python
# On crée une colonne genre
dfTrain['Gender'] = dfTrain['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
# On fait ensuite une copie de la colonne age:
dfTrain['AgeFill'] = dfTrain['Age']

dfTrain[ dfTrain['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

Calculons maintenant la médiane par genre et par classe. Nous allons avoir besoin de numpy pour cela. 

```python
import numpy as np
```

```python
median_ages = np.zeros((2,3))
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = dfTrain[(dfTrain['Gender'] == i) & \
                              (dfTrain['Pclass'] == j+1)]['Age'].dropna().median()
median_ages
```

```python
Remplissons maintenant la colonne AgeFill avec nos données. 
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        dfTrain.loc[ (dfTrain.Age.isnull()) & (dfTrain.Gender == i) & 
                (dfTrain.Pclass == j+1),\
                'AgeFill'] = median_ages[i,j]
dfTrain[ dfTrain['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```
