# Nettoyage des données


En observant les données nous constatons que des âges ne sont pas renseignés. Nous allons donc créer une colonne de copie de l'âge puis nous remplacerons les valeurs par la médiane. 


```python
dfTrain.Age.mean()
dfTrain[['Sex', 'Pclass', 'Age']]
dfTrain[dfTrain['Age'] > 60]
dfTrain[dfTrain['Age'] > 60][['Sex', 'Pclass', 'Age', 'Survived']]
dfTrain[dfTrain['Age'].isnull()][['Sex', 'Pclass', 'Age']]
```

```python
dfTrain.Age.mean()
```

```python
dfTrain.Age.median()
```

L'age médian est de 28 ans alors que la moyenne est de 29.69
On définit une colonne genre. 
On calcule la médiane par classe et par genre

```python
median_ages = np.zeros((2,3))
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = dfTrain[(dfTrain['Gender'] == i) & \
                              (dfTrain['Pclass'] == j+1)]['Age'].dropna().median()
median_ages
```

```python
# On crée une colonne genre
dfTrain['Gender'] = dfTrain['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
# On fait ensuite une copie de la colonne age:
dfTrain['AgeFill'] = dfTrain['Age']

dfTrain[ dfTrain['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

On remplit la colonne des âges en la remplaçant par la médiane

```python
# Et puis on le rempli :
for i in range(0, 2):
    for j in range(0, 3):
        dfTrain.loc[ (dfTrain.Age.isnull()) & (dfTrain.Gender == i) & 
                (dfTrain.Pclass == j+1),\
                'AgeFill'] = median_ages[i,j]
dfTrain[ dfTrain['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
dfTrain['AgeIsNull'] = pd.isnull(dfTrain.Age).astype(int)
dfTrain.head(10)
```
